```python
num_students = int(input('Toplam kaç öğrenci var?  '))
num_test_scores = int(input("Öğrenci başı kaç test var?  "))

for std in range(num_students):
    total=0.0
    
    print('Öğrenci numarası',std+1)
    print("//////////////////////")
    
    for tn in range(num_test_scores):
        print("Test numarası", tn+1, end='')
        score=float(input(": "))
        
        total += score
        
    avg = total / num_test_scores
        
    print('Öğrenci {} için ort is:'.format(std+1), avg)
    
      
    
```
